```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('SLH_100.csv')
```

```python
df['Label']=df['Label'].apply(lambda x: 1 if x > 0.5 else 0 )
```

```python
df
```

```python
df['fn'] = df.ID.apply(lambda x: '_'.join(x.split('_')[:2]) + '.png')
df.head(5)
```

```python
df
```

```python
df.columns = ['ID', 'probability', 'fn']
```

```python
df
```

```python
df['label'] = df.ID.apply(lambda x: x.split('_')[-1])
```

```python
df
```

```python
df.drop_duplicates('ID', inplace=True)
```

```python
df
```

```python
from collections import defaultdict

d = defaultdict(list)
for fn in df.fn.unique(): d[fn]

for tup in df.itertuples():
    if tup.probability: d[tup.fn].append(tup.label)
```

```python
ks, vs = [], []

for k, v in d.items():
    ks.append(k), vs.append(' '.join(v))
```

```python
pd.DataFrame(data={'fn': ks, 'labels': vs}).to_csv('SLH_500_STRINGS.csv', index=False)
```
